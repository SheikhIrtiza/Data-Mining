```python
#we can replace 0's with NaN
#The records that have 0's in columns Glucose, Blood Pressure, Skin Thickness, Insulin and BMI will be replaced with NaN

from numpy import nan

data['Glucose']=data['Glucose'].replace(0,np.nan)
```

```python
data['BloodPressure']=data['BloodPressure'].replace(0,np.nan)
```

```python
data['SkinThickness']=data['SkinThickness'].replace(0,np.nan)
```

```python
data['Insulin']=data['Insulin'].replace(0,np.nan)
```

```python
data['BMI']=data['BMI'].replace(0,np.nan)
```

```python
#Lets check if the above code has worked
data.head()
#All the 0's have been replaced by NaN
```

```python
data.describe()
```

```python
#Count if NaN values in the dataset
print(data.isnull().sum())
```

```python
#As inferred colums glucose , bp, st,insulin and BMI have missing data.
#1. Glucose, BP and BMI have less number of missing data
#2. but skin thickness and insulin have a very high amount of missing data.
 ###Removing these missing data from the data set will result in information loss and it is not 
 ###advisible to remove  these records since we have only 768 records. Hence we will import these
 ### missing values with the median of their respective colums, since median is at least
 ### affectd by Outliers.
```

```python
data.median()
```

```python
#importong mising values with their respectove columns median
data.fillna(data.median(), inplace=True)
```
